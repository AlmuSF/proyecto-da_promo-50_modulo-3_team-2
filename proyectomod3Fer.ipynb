```python
# borrar columnas 

porcentaje_nulos = df.isna().mean() * 100
print(porcentaje_nulos.sort_values(ascending=False))
df = df.drop(columns=['Unnamed: 0', "employeecount", "numberchildren",'over18', 'hourlyrate', 'yearsincurrentrole','roledepartament','sameasmonthlyincome', 'educationfield', 'department'])
df.columns
```

```python
# quitar duplicados

df = df.drop_duplicates(subset=["employeenumber"], keep="first")
df
```

```python
# Reemplazamos el valor 'Married' por 'Marreid' para unificar valores.
df['maritalstatus'] = df['maritalstatus'].replace("Marreid", "Married")
```

```python
#Compruebo que se ha reemplazado.
df['maritalstatus'].unique()
```

```python
# Reemplazamos el valor '0' por 'M' y el 1 por 'F' para unificar valores.
def reemplazar_valores(df, gender):
    df[gender] = df[gender].replace({0:"M", 1: "F"})
    return df
```

```python
df = reemplazar_valores(df,'gender')
```

```python
df.dtypes
```

```python
df['standardhours'] = df['standardhours'].astype(str) 
print(df.dtypes)
```

```python
df["attrition"] = df["attrition"].str.lower()
df["businesstravel"] = df["businesstravel"].str.lower()
df["jobrole"] = df["jobrole"].str.lower()
df["maritalstatus"] = df["maritalstatus"].str.lower()
df["overtime"] = df["overtime"].str.lower()
df["standardhours"] = df["standardhours"].str.lower()
df["remotework"] = df["remotework"].str.lower()
```

```python
# Aplicar conversiÃ³n de texto a nÃºmero en la columna 'age'
df["age"] = df["age"].apply(lambda x: w2n.word_to_num(x) if isinstance(x, str) else x)
print(df)
```

```python
# Columnas numÃ©ricas que venÃ­an como texto con coma en lugar de punto
columnas_con_coma = ['performancerating', 'totalworkingyears', 'worklifebalance']

# Columnas tipo object que deben convertirse a numÃ©ricas (int o float)
columnas_a_convertir = [
    'totalworkingyears', 'performancerating', 'standardhours',
    'stockoptionlevel', 'trainingtimeslastyear', 'percentsalaryhike',
    'worklifebalance'
]
```

```python
# Eliminar sÃ­mbolo de dÃ³lar si aparece
def eliminar_dolar(df):
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].str.replace('$', '', regex=False)
    return df
```

```python
df = eliminar_dolar(df)
```

```python
# Eliminar espacios en columnas tipo texto
def eliminar_espacios(df):
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].str.strip()
    return df
```

```python
# Reemplazar ',' por '.' y convertir a numÃ©rico
def limpiar_columnas_numericas_con_coma(df, columnas):
    for col in columnas:
        if col in df.columns and df[col].dtype == 'object':
            df[col] = df[col].str.replace(',', '.', regex=False)
    return df
```

```python
# Convertir columnas a numÃ©rico: int si no hay nulos, float si los hay
def convertir_columnas_a_entero_si_es_posible(df, columnas):
    for col in columnas:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            if df[col].isnull().sum() == 0:
                df[col] = df[col].astype('int')
    return df
```

```python
# Redondear floats a 2 decimales
def redondear_floats(df):
    float_cols = df.select_dtypes(include='float')
    df[float_cols.columns] = float_cols.round(2)
    return df

```

```python
# FunciÃ³n principal de limpieza bÃ¡sica
def limpieza_basica_eda(df):
    df = eliminar_dolar(df)
    df = eliminar_espacios(df)
    df = limpiar_columnas_numericas_con_coma(df, columnas_con_coma)
    df = convertir_columnas_a_entero_si_es_posible(df, columnas_a_convertir)
    df = redondear_floats(df)
    return df
```

```python
# AplicaciÃ³n de la limpieza al DataFrame
df = limpieza_basica_eda(df)

# Revisar resultados
df.info()
```
